```python
from joblib import dump, load
import pandas as pd
import json

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
import numpy as np
import matplotlib as mpl
```

```python
df = pd.read_parquet('./data/data_df_hcm_v3.parquet')
target_df = pd.read_parquet('./data/target_df_hcm_v3.parquet')

load_model = load("./model/hcm/cat/v3/model.joblib")
FS = json.load(open('./data/hcm_v3.json', 'r'))
cat_cols = FS['cat_cols']
num_cols = FS['num_cols']
all_cols = cat_cols + num_cols

df = df[all_cols]
```

```python
df['target'] = target_df['target']
df['event_timestamp'] = target_df['event_timestamp']

df = df.reset_index(drop = True)
```
